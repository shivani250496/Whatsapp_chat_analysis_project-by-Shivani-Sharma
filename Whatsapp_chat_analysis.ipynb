```python
def startsWithDateAndTime(s):
    pattern = r"[\d]{1,2}/[\d]{1,2}/[\d]{4}" #, r"([0-9]+):([0-9]+)[ ]?(AM|PM|am|pm)? -" # [\d]{1,2}/[\d]{1,2}/[\d]{4} ^([0-9]+)(\/)([0-9]+)(\/)([0-9]+)
    result = re.findall(pattern, s)
    print("result :", result)
    if result:
        return True
    else:
        False



```

```python
def FindAuthor(s):
    s=s.split(":")
    if len(s)==2:
        return True
    else:
        return False
```

```python
def getDataPoint(line):
    splitLine = line.split(' - ')
    dateTime = splitLine[0]
    date, time = dateTime.split(', ')
    message = ' '.join(splitLine[1:])
    if FindAuthor(message):
        splitMessage = message.split(': ')
        author = splitMessage[0]
        message = ' '.join(splitMessage[1:])
    else:
        author = None
    return date, time, author, message
```

```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Example DataFrame setup for demonstration
df = pd.DataFrame({
    'Date': pd.to_datetime([]),
    'Time': [],
    'Author': [],
    'Message': [],
    'emoji': [],
    'urlcount': []
})

# Add your actual data to the DataFrame here
# df = pd.read_csv('your_whatsapp_chat.csv')

# List of authors (example)
authors = ['Author1', 'Author2', 'Author3']

# Function to generate stats and word clouds for each author
def print_stats_and_wordcloud(df, authors):
    for author in authors:
        req_df = df[df['Author'] == author]
        print(f"Stats of {author}")

        total_messages = req_df.shape[0]
        print('Messages Sent', total_messages)

        if total_messages > 0:
            # Calculate words per message
            words_per_message = (np.sum(req_df['Message'].str.split().str.len())) / total_messages
            print('Words per message', words_per_message)

            # Calculate media messages
            media_messages = req_df[req_df['Message'] == '<Media omitted>'].shape[0]
            print('Media Messages', media_messages)

            # Calculate emojis
            emojis = req_df['emoji'].str.len().sum()
            print('Emojis Sent', emojis)

            # Calculate links
            links = req_df['urlcount'].sum()
            print('Links Sent', links)

            # Generate word cloud
            text = ' '.join(req_df['Message'])
            if text.strip():  # Check if text is not empty
                stopwords = set(STOPWORDS)
                wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

                # Display the generated image:
                plt.figure(figsize=(10, 5))
                plt.imshow(wordcloud, interpolation='bilinear')
                plt.axis("off")
                plt.title(f"Word Cloud for {author}")
                plt.show()
            else:
                print(f'No text available to generate word cloud for {author}.')
        else:
            print(f'No messages sent by {author}.')

# Call the function with the DataFrame and list of authors
print_stats_and_wordcloud(df, authors)
```
